# df_names : nettoyage dataset name basics


```python
df_names = pd.read_csv('https://datasets.imdbws.com/name.basics.tsv.gz', sep="\t")
```

```python
# Remplacer les valeurs \N par nul
df_names = df_names.replace("\\N",np.nan)
```

```python
# Supprimer les valeurs si valeurs colonne primary name, birthday, deathyear,primaryProfession et knownForTitles sont nulles
cast = df_names.dropna(subset=["primaryName","birthYear","deathYear","primaryProfession","knownForTitles"],axis=0, how="all")
```

```python
# Supprimer si "primaryProfession" est nulles
df_names.dropna(subset=["primaryProfession"], inplace=True)
```

```python
# Vérifier les valeurs uniques de la colonne "PrimaryProfession"
df_names["primaryProfession"].isnull().sum()
```

```python
#Supprimer si dates de naissance et de décès non spécifié
df_names.dropna(subset=["birthYear","deathYear"],how="all",inplace=True)
```

```python
# Supprimer les lignes des cast n'étant pas un director,actor et actress

# Filtrer les lignes où la colonne "primaryProfession" contient les valeurs "actor", "actress" ou "director"
df_cast = df_names[df_names["primaryProfession"].str.contains("actor|actress|director")]

# Afficher les lignes filtrées
display(df_cast)

```

```python
df_cast.describe()
```

```python
# Créer df contenant tous les directeurs
df_director = df_cast[df_cast["primaryProfession"].str.contains("director")]
df_actor = df_cast[df_cast["primaryProfession"].str.contains("actor|actress")]
```

```python
display(df_actor)
```

```python
# Création DF des acteurs et actrices décédés 
df_actor_death = df_actor[df_actor["deathYear"].notnull()]

# Création DF des acteurs et actrices vivants
df_actor_alive = df_actor[df_actor["deathYear"].isna()]
```

```python
# Renommer la colonne "deathYear" en "Age"
df_actor_alive.rename(columns={"deathYear": "Age"}, inplace=True)

# Définir l'année actuelle
year_current = datetime.datetime.now().year

# Convertir la colonne "birthYear" en entier
df_actor_alive["birthYear"] = df_actor_alive["birthYear"].astype(int)

# Calculer l'âge en soustrayant "birthYear" de l'année actuelle
df_actor_alive["Age"] = year_current - df_actor_alive["birthYear"]
```

```python
# Définir l'age des acteurs actuellement vivants
df_actor_alive
```

```python
# Anomalie détecter, des dates de naissance n'ont pas été renseigné alors que l'année de décès oui = on supprime
df_actor_death["birthYear"].isnull().sum()
```

```python
# On supprime les lignes qui n'ont pas renseigné leur date de naissance alors que de décès oui
df_actor_death = df_actor_death.dropna(subset=["birthYear"])
```

```python
# Convertir les colonnes "birthYear" et "deathYear" en entiers
df_actor_death[["birthYear", "deathYear"]] = df_actor_death[["birthYear", "deathYear"]].astype(int)

# Calculer l'âge au décès en soustrayant "birthYear" de "deathYear"
df_actor_death["Age Décès"] = df_actor_death["deathYear"] - df_actor_death["birthYear"]
```

# df_titles crew : Nettoyage dataset Title Crew

```python
# Importer Dataset title crew
df_titles_crews = pd.read_csv("https://datasets.imdbws.com/title.crew.tsv.gz",sep="\t")
```

```python
# Remplacer les valeurs \N par nul
df_titles_crews.replace("\\N",np.nan,inplace=True)
```

```python
# Afficher les premières lignes de df_names
df_titles_crews.head()
```

```python
#Supprimer la colonne writers
df_titles_crews = df_titles_crews.drop("writers",axis=1)
```

```python
# Afficher df_title avec MAJ
display(df_titles_crews)
```

```python
# On supprime toutes les lignes directors nulles
df_titles_crews = df_titles_crews.dropna(subset =["directors"])
```

```python
# Toutes les lignes directors nulles ont bien été supprimées
df_titles_crews["directors"].isnull().sum()
```

```python
# Transformer la ligne directors en liste
df_titles_crews["directors"] = df_titles_crews["directors"].map(lambda x: x.split(","))
```
