```python
import pandas as pd
import numpy as np
```

```python
df_movie = pd.read_pickle("df_movieOK.pkl")
df_character = pd.read_pickle("df_moviecharacters.pkl")
```

```python
# Supprimer la colonne imdb_id 
df_movie = df_movie.drop(columns="imdb_id")
```

```python
df_movie = df_movie.merge(df_character, on="tconst", how="inner")
```

```python
df_movie.head()
```

```python
# Supprimer les colonnes inutiles suite au merge
df_movie = df_movie.drop(columns=["averageRating_y","numVotes_y","isAdult_y","startYear_y","runtimeMinutes_y","genres_y","title_y"])
```

```python
# Renommer les colonnes

df_movie = df_movie.rename(columns={"averageRating_x":"averageRating",
                                    "numVotes_x":"numVotes",
                                    "isAdult_x":"isAdult",
                                    "startYear_x":"startYear",
                                    "runtimeMinutes_x":"runtimeMinutes",
                                    "genres_x":"genres",
                                    "title_x":"title",
                                    "dict_characters":"casting"})
df_movie.head()
```

```python
df_movie.to_pickle("df_movieOK.pkl")
```
