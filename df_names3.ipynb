# EXPLORATION ET NETTOYAGE DE LA TABLE DF_NAMES

```python
df_names = pd.read_csv('https://datasets.imdbws.com/name.basics.tsv.gz', sep="\t")
```

```python
# Remplacer les valeurs \N par nul
df_names = df_names.replace("\\N",np.nan)
```

```python
# Supprimer les valeurs si valeurs colonne primary name, birthday, deathyear,primaryProfession et knownForTitles sont nulles
cast = df_names.dropna(subset=["primaryName","birthYear","deathYear","primaryProfession","knownForTitles"],axis=0, how="all")
```

```python
# Supprimer si "primaryProfession" est nulles
df_names.dropna(subset=["primaryProfession"], inplace=True)
```

```python
df_names.shape
```

```python
# Vérifier les valeurs uniques de la colonne "PrimaryProfession"
df_names["primaryProfession"].isnull().sum()
```

```python
#Supprimer si dates de naissance et de décès non spécifié
df_names.dropna(subset=["birthYear","deathYear"],how="all",inplace=True)
```

```python
# Supprimer les lignes des cast n'étant pas un director,actor et actress

# Filtrer les lignes où la colonne "primaryProfession" contient les valeurs "actor", "actress" ou "director"
df_cast = df_names[df_names["primaryProfession"].str.contains("actor|actress|director")]

# Afficher les lignes filtrées
display(df_cast)

```

```python
df_cast.describe()
```

```python
# Supprimer les lignes où birthYear ou knownForTitles n'est pas renseigné
df_cast = df_cast.dropna(subset=["birthYear"], axis=0)
df_cast = df_cast.dropna(subset=["knownForTitles"], axis=0)
```

```python
# Nombre de nulls dans la colonne birthyear qui bloque le calcul age
df_cast["birthYear"].isnull().sum()
```

```python
# Créer la colonne Alive pour savoir s'ils sont vivants ou décédés
df_cast["alive"] = df_cast["deathYear"].isnull().astype(int)
```

```python
# Convertir les types 
# De la colonne birthYear en int
df_cast["birthYear"] = df_cast["birthYear"].astype(int)

# De la colonne alive en booléen
df_cast["alive"] = df_cast["alive"].astype(bool)
```

```python
# Création de la colonne age selon acteur décédé ou vivant

def calcul_age(ligne):

    year_current = datetime.datetime.now().year             # Définir année actuel
    if ligne["alive"] == 1:                                 # Condition si la colonne alive de la ligne est égal à 1 donc est vivant 
        return year_current - ligne["birthYear"]
    else:
        death_year = int(ligne["deathYear"]) if pd.notnull(ligne["deathYear"]) else None
        return death_year - ligne["birthYear"] if death_year else None

# Appliquer la fonction calcul_age à chaque ligne du DataFrame
df_cast["age"] = df_cast.apply(calcul_age, axis=1)
```

```python
# Supprimer les "_" de la colonne primaryProfession
df_cast["primaryProfession"] = df_cast["primaryProfession"].map(lambda x : x.replace("_"," "))
```

```python
# Transformer la colonne knowforTitles et primaryProfession en liste
df_cast['primaryProfession'] = df_cast['primaryProfession'].astype(str)
df_cast["primaryProfession"] = df_cast["primaryProfession"].map(lambda x : x.split(","))
df_cast['knownForTitles'] = df_cast['knownForTitles'].astype(str)
df_cast["knownForTitles"] = df_cast["knownForTitles"].map(lambda x : x.split(","))
```

```python
# Nettoyer la liste des métiers
metier_suppr = ["music department","miscellaneous","archive footage","soundtrack","archive footage"
                ,"publicist","casting director","make up department","talent agent","casting director","manager","legal","casting department",
                "camera department","editor","location management","archive sound","costume department","production designer","set decorator","transportation department",
                "animation department","art department","script department","sound department","costume designer","visual effects","visual effects","special effects",
                "accountant","stunts","editorial department","cinematographer"]


# Appliquer la fonction pour supprimer les métiers spécifiés de chaque liste
df_cast['primaryProfession'] = df_cast['primaryProfession'].map(lambda liste: [valeur for valeur in liste if valeur not in metier_suppr])
```

```python
# Définition de la fonction pour joindre les valeurs
def join_values(group):
    return ','.join(group)

# Groupement par "primaryProfession" et application de la fonction join_values
list_metiers = df_cast["primaryProfession"].map(lambda x : ",".join(x))
list_metiers = list_metiers.drop_duplicates()

# Groupement par "knownForTitles" et application de la fonction join_values
list_films = df_cast["knownForTitles"].map(lambda x : ",".join(x))
list_films = list_metiers.drop_duplicates()

# Affichage des valeurs uniques et leur compte
print(list_metiers.count())
print(list_metiers.value_counts())

print(list_films.count())
print(list_films.value_counts())
```

```python
grouped = df_cast.groupby(['primaryName', 'birthYear'])

def merge_lists(lists):
    merged_list = []
    for item in lists:
        merged_list.extend(item) # Ajouter à une liste existante pour éviter de créer une nouvelle liste pour chaque élément ajouté.
    return list(set(merged_list)) # On utilise le type set pour supprimer les doublons pour ensuite le reconvertir en liste

df_cast = grouped.agg({
    'nconst' : 'first', # Prendre la première valeur pour deathYear (puisqu'elles sont identiques pour chaque groupe)
    'deathYear': 'first',  # Prendre la première valeur pour deathYear (puisqu'elles sont identiques pour chaque groupe)
    'primaryProfession': merge_lists,
    'knownForTitles': merge_lists,
    'alive': 'first',  # Prendre la première valeur pour alive (puisqu'elles sont identiques pour chaque groupe)
    'age': 'first'  # Prendre la première valeur pour age (puisqu'elles sont identiques pour chaque groupe)
}).reset_index()
```
